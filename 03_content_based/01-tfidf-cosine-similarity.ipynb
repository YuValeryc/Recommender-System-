```python
file_path = '/kaggle/input/the-movies-dataset/movies_metadata.csv'

df = pd.read_csv(file_path, low_memory=False)

print("Shape of origin data:", df.shape)
display(df.head())
print("Infomation of data", df.info())
```

```python
df = df[['id', 'title', 'overview']]
display(df.iloc[:100])
df.dropna(subset=['overview', 'title'], inplace=True)

df['id'] = pd.to_numeric(df['id'], errors='coerce')
df.dropna(subset=['id'], inplace=True)
df['id'] = df['id'].astype(int)

df.drop_duplicates(subset='title', inplace=True)

df.reset_index(drop=True, inplace=True)

print("Shape of df", df.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english', min_df=3)
print("Matrix TFIDF is creating...")
tfidf_matrix = tfidf.fit_transform(df['overview'])
print("Done")
print(tfidf_matrix)
```

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
print("Shape of sim matrix:", cosine_sim.shape)
print("\nSome sample of matrix (5x5):")
print(cosine_sim[:5, :5])
```

```python
indices = pd.Series(df.index, index=df['title'])

def get_recommendations(title, cosine_sim=cosine_sim, data=df, indices=indices):
    try:
        idx = indices[title]
    except KeyError:
        print(f"The movie {title} is not exist in data")

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]

    return data['title'].iloc[movie_indices]

print("--- Gợi ý cho 'The Dark Knight Rises' ---")
display(get_recommendations('The Dark Knight Rises'))

print("\n--- Gợi ý cho 'The Matrix' ---")
display(get_recommendations('The Matrix'))

print("\n--- Gợi ý cho 'Toy Story' ---")
display(get_recommendations('Toy Story'))
```
